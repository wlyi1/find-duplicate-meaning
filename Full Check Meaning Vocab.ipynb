```python
import pandas as pd
import numpy as np
df = pd.read_excel(r'C:\Users\wlyi\Documents\English\vocabe.xlsx')
pro = df.values.tolist()
vocab = np.ravel(pro)

def get_duplicate(vocab):
    
    dict_of_duplicate = dict()
    index = 0
    # lakukan iterasi element dalam list dan masukkan indeksnya
    for elem in vocab:
        # jika element ada di dict, terus masukkan index dan tambahkan jumlah frekeuensinya
        if elem in dict_of_duplicate:
            dict_of_duplicate[elem][0] += 1
            dict_of_duplicate[elem][1].append(index)
        else:
            # tambah key-value baru 
            dict_of_duplicate[elem] = [1, [index]]
        index += 1    
 
    dict_of_duplicate = { key:value for key, value in dict_of_duplicate.items() if value[0] > 1}
    return dict_of_duplicate

dict_of_duplicate = get_duplicate(vocab)
for key, value in dict_of_duplicate.items():
       print('Element, ',key , ', Repeated Count,',value[0] , ',Index Positions, ',value[1])
        
```

```python
del dict_of_duplicate['nan']
```
